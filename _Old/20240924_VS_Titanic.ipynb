```python
X = train_df.drop(columns=['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'])
X.head()
```

```python
# Dictionnaire de mappage pour convertir les valeurs catégorielles en entiers
mapping = {'S': 1, 'C': 2, 'Q': 3}
X['Embarked'] = X['Embarked'].map(mapping)
X.head()
```

```python
# Transformer la colonne 'Sex' en une colonne de booléens
sex_mapping = {'male': True, 'female': False}
X['Sex'] = X['Sex'].map(sex_mapping)
X.head()
```

```python
Y = train_df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, train_size=42)

X_train.head()
```
